## Bar Graph :

```python
#Python code for the below bar graph

l1=np.array(df['Current Assets/Total Assets'])
l2=np.array(df['Current Liability to Liability'])

#claculating a new value using given cols
CurrentRatio=[x/y if y!=0 else y for x,y in zip(l1,l2)]
crs =pd.Series(CurrentRatio)

#three catagories

unstable = crs[crs < 0.5].count()
stable=crs[(crs >= 0.5) & (crs <1)].count()
ideal = crs[(crs >=1 )].count()

#figure size
plt.figure(figsize =(6,6))
plt.style.use('ggplot')

#labels

labels=['unstable','stable','ideal']
values=[unstable,stable,ideal]
bars=plt.bar(labels ,values , width=0.4 ,label='companys count',color='blue' ,alpha=0.7)
plt.title('ShortTerm Financial Stability (Based On Current Ratio)')

plt.yticks(np.arange(0,6000,500))
plt.xlabel('State Of Stability')
plt.ylabel('Number Of Companys')
plt.legend()
plt.tight_layout()
plt.show()

#A company is said to be 
#->State of company  ->Current Ratio
#->Unstable          -> < 0.5
#->stable            -> > 0.5 & < 1
#->ideal             -> == 1

#Most of the companys are stabel & only few are bankrupted
```

#  Model

```python
# 1) Importing Classes Of Models That I am going to use.

from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC

from sklearn.ensemble import RandomForestClassifier

# KNN & SVC are Distance Based Algos
# LogisticRegression is a Gradient-Based Algo
# I Am going to apply PCA (Dimentionality Reduction)
# For This Three Algos Scaling is Crucial
# So , I am going to Scale (normalize) my dataset (By This Mean=0 & STD=1 for every numeric Column)
# Before Doing This We Need To Split Our Dataset Into Training & Testing Sets.

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score
```

```python
# Preprocessing My DataSet

def preprocess_input(df):
    df=df.copy()
    
    #Split df into X and y
    y=df['Bankrupt?']
    X=df.drop('Bankrupt?',axis=1)
    
    # Train Test Split
    X_train , X_test , y_train , y_test = train_test_split(X, y, train_size=0.7, shuffle=True, random_state=1)
    
    #Scaleing DataSet
    scaler = StandardScaler()
    scaler.fit(X_train) # Notice unsupervised learning (target vector is not given).
    X_train = pd.DataFrame(scaler.transform(X_train), index=X_train.index ,columns = X_train.columns)
    X_test = pd.DataFrame(scaler.transform(X_test), index=X_test.index ,columns = X_test.columns)
    
    return X_train , X_test , y_train , y_test;
```

```python
X_train , X_test , y_train , y_test = preprocess_input(df)

#Data is scaled and splited.
X_train.head()
```

```python
X_test.head()
```

```python
# 2)Creating All The Models.

models ={
     'LogisticRegression':LogisticRegression(),
          'Decision Tree':DecisionTreeClassifier(),
 'Support Vector Machine':SVC(),
          'Random Forest':RandomForestClassifier()
}
```

```python
# 3) Fitting Data To Our Models (Training Models).

for modelname,model in models.items():
    model.fit(X_train,y_train)
    print(modelname +": trained.")
```

```python
# 4) Testing Our Models Accuracy.
results=[]

for name ,model in models.items():
    result = model.score(X_test,y_test)
    results.append(result)
    print(name +":  {:.2f}%".format(result *100))
    
# By This We Can Conclude That 
# =>SVM Model is Giving the Best Results.96.60
```

# LogisticRegression:  96.43%

```python
#Best Hyper Perameter selection.(Applying RandomizedSearchCV)


from sklearn.model_selection import RandomizedSearchCV
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report



# Step 1: Define the parameter grid
param_grid = {
    'C': np.logspace(-3, 2, 6),  # [0.001, 0.01, 0.1, 1, 10, 100]
    'gamma': np.logspace(-3, 2, 6),  # [0.001, 0.01, 0.1, 1, 10, 100]
    'kernel': ['linear', 'rbf', 'poly', 'sigmoid']
}

# Step 4: Initialize the SVC and RandomizedSearchCV
svc = SVC()
random_search = RandomizedSearchCV(svc, param_distributions=param_grid, n_iter=50, cv=5, scoring='accuracy', random_state=42)

random_search.fit(X_train, y_train)

# Step 6: Evaluate the results
print("Best Parameters:", random_search.best_params_)
print("Best Score:", random_search.best_score_)

#This Process takes lots of time to complete
```

# Best Parameters: {'kernel': 'linear', 'gamma': 0.1, 'C': 0.1}
# Best Score: 0.9702490478228896

# Metrics

```python
# Optionally, predict on the test set and evaluate

y_pred = random_search.best_estimator_.predict(X_test)

#Classification report of logistic model

print(classification_report(y_test, y_pred))
best_model=random_search.best_estimator_
```

## Applying PCA (Dimentionality Reduction) For Enhancing Model :

```python
# importing Required Packages.
# Dimentionality Reduction is under Unsupervised learning.

from sklearn.decomposition import PCA
n_components=10

# Now Decomposing 95 column dataset into 10 column dataset.

PCA_model = PCA(n_components)

PCA_model.fit(X_train)

columns = ['PCA{0}'.format(i) for i in range(n_components)]


X_train_PCA = pd.DataFrame(PCA_model.transform(X_train),index = X_train.index , columns = columns )
X_test_PCA = pd.DataFrame(PCA_model.transform(X_test),index = X_test.index , columns = columns )

X_train_PCA.head()

# Now we successfully reduced 95D data into 10D dataset Without Loosing the essential features of the dataset.
```
